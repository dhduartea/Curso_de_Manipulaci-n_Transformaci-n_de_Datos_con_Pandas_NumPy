## Group_by
Permite agrupar datos en función de los demás. Es decir, hacer el análisis del DataFrame en función de una de las columnas.



- Agrupar por Author y mostrar el conteo de los datos de las demás columnas


```python
df_salaries.groupby('job_title').count()
```

- funciona similar al value_counts(), pero en este caso nos entrega todos los datos

```python
df_salaries.value_counts("job_title")
```


- Agrupar por Author y mostrar la mediana de los datos de las demás columnas


```python
df_salaries.groupby('job_title').median()
```


La columna job_title, en los casos anteriores, pasa a ser el índice.



Podemos usar loc y acceder a un dato específico del DataFrame. Agrupar por autor y mostrar la suma de los valores de las demás columnas para Data Scientist


```python
df_salaries.groupby('job_title').sum().loc['Data Scientist']
```

- Agrupar por job_title y mostrar la suma de los valores de las demás columnas. Colocar los índices que el DataFrame trae por defecto


```python
df_salaries.groupby('job_title').sum().reset_index()
```


- La función agg() permite aplicar varias funciones al DataFrame una vez agrupado según una columna específica. Agrupar por job_title y mostrar el mínimo y máximo de las demás columnas

```python
df_salaries.groupby('job_title').agg(['min','max'])#muestra cada columna dividida en dos: min y max. Estas contienen los valores máximo y mínimo de la columna para cada job_title 
```


- Agrupar por job_title, obtener el mínimo y máximo de la columna ‘salary_in_usd’ y obtener la media de los valores de la columna ‘salary’


```python
df_salaries.groupby('job_title').agg({'salary_in_usd':['min','max'], 'salary':'mean'})
```


Agrupar por ‘job_title - work_year’ y obtener la mediana de  los valores de las demás columnas


```python
df_salaries.groupby(['job_title','work_year']).median()
```

## Uniones


### Concat
En esta ocasión vamos a crear un DataFrame nuevo.


```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2','A3'],
        'B':['B0', 'B1', 'B2','B3'],
	'C':['C0', 'C1', 'C2','C3'],
	'D':['D0', 'D1', 'D2','D3']})


df2 = pd.DataFrame({'A':['A4', 'A5', 'A6','A7'],
	'B':['B4', 'B5', 'B6','B7'],
	'C':['C4', 'C5', 'C6','C7'],
	'D':['D4', 'D5', 'D6','D7']})
```

- Concatenar los DataFrames


```python
pd.concat([df1,df2])
```

- Corregir los índices


```python
pd.concat([df1,df2], ignore_index= True)
```

- Por axis 1 (columnas)


```python
pd.concat([df1,df2], axis = 1)
```

### Merge


#### MERGE 1
Creamos DataFrame


```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
        'A' : ['A0', 'A1', 'A2','A3'],
        'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
        'C' : ['C0', 'C1', 'C2','C3'],
        'D': ['D0', 'D1', 'D2','D3']})
        
```

- Unir el DataFrame Der a Izq


```python
izq.merge(der)
```

#### MERGE 2


```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
        'A' : ['A0', 'A1', 'A2','A3'],
        'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
        'C' : ['C0', 'C1', 'C2','C3'],
        'D': ['D0', 'D1', 'D2','D3']})
```


- Hay diferencias entre algunas columnas, por esa razón hay que separarlos de esta manera:


```python
izq.merge(der, left_on = 'key', right_on='key_2')
```

#### MERGE 3


```python

izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
        'A' : ['A0', 'A1', 'A2','A3'],
        'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
        'C' : ['C0', 'C1', 'C2','C3'],
        'D': ['D0', 'D1', 'D2','D3']})
```

- Si tenemos un NaNen nuestro DataFrame, pandas no lo detectará como un mach. Se soluciona con How, dando así, una preferencia.


```python
izq.merge(der, left_on = 'key', right_on='key_2')
```

```python
izq.merge(der, left_on = 'key', right_on='key_2',how='left')
```

```python
izq.merge(der, left_on = 'key', right_on='key_2',how='inner')
```

```python
izq.merge(der, left_on = 'key', right_on='key_2',how='right')
```

### Join
Es otra herramienta para hacer exactamente lo mismo, una combinación. La diferencia es que join va a ir a los índices y no a columnas específicas.


```python

izq = pd.DataFrame({'A': ['A0','A1','A2'],
      'B':['B0','B1','B2']},
      index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
      'D':['D0','D1','D2']},
      index=['k0','k2','k3']) 
```

- Combinamos izq con der


```python
izq.join(der)
```

```python
izq.join(der, how = 'left')
```

- Traer todos los datos aunque no hagan match.


```python
izq.join(der, how = 'outer')
```

```python
izq.join(der, how = 'inner')
```

```python
der.join(izq, how = 'outer')
```

### Apply
 Es un comando muy poderoso que nos deja aplicar funciones a nuestro DataFrame


- Creamos unos DataFrame habitual

```python
df_salaries.head(2)
```

```python
def two_times(value):
    return value * 2
```

- Lo aplicamos a la columna de salary

```python
df_salaries["salary"].apply(two_times) #Se multiplica por 2 todos los valores de la columna
```

- Podemos guardarlo en una columna nueva

```python
df_salaries["SalaryDoble"]=df_salaries["salary"].apply(two_times)
```

```python
df_salaries.head(2)
```

- Se pueden crear lambda functions

```python
df_salaries['salary2'] =df_salaries['salary'].apply(lambda x: x* 3)
```

- Apply en varias columnas con condiciones, hay que especificar a que los vamos a aplicar (filas o columnas)

```python
df_salaries["data_scientist"]=df_salaries.apply(lambda x: x['salary'] * 2 if x['job_title'] == 'Data Scientist' else x['salary'], axis = 1)
# Multiplica por 2 a los datos que cumplan la condición 
```
