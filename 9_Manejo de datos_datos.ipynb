## Agregar o eliminar datos con Pandas

Muchas ocasiones necesitamos agregar, eliminar o separar datos y pandas nos ofrece varias funciones para que este proceso se vuelva mucho más sencillo.

Muestra las primeras 5 líneas del DataFrame


```python
import numpy as np
import pandas as pd
```

```python
df_salaries=pd.read_csv("ds_salaries.csv",index_col=0,sep=",")
df_salaries
```

```python
df_salaries.head()  #muestra las primeras 5 lineas del dataFrame
```

Eliminar columnas de la salida pero no del DataFrame, si se quieren eliminar del dataframe original, se debe asignar **True** a la propiedad *inplace*


```python
df_salaries.drop('employee_residence', axis=1).head() #elimina la columna employee_residence de la salida pero no del dataFrame
```

Eliminar una columna


```python
del df_salaries['company_size']  #elimina la columna company_size del dataFrame
df_salaries
```

Eliminar filas


```python
df_salaries.drop(0, axis=0) #elimina la fila 0 del dataFrame
```

Eliminar un conjunto de filas mediante una lista


```python
df_salaries.drop([0,1,2], axis=0) #elimina las filas 0, 1 y 2 del dataFrame
```

Elimina un conjunto de filas mediante un rango


```python
df_salaries.drop(range(0,10), axis=0) #elimina las primeras 10 filas del dataFrame
```

Agregar una nueva columna con valores Nan


```python
df_salaries['Nueva_columna'] = pd.NA #Crea una nueva columna con el nombre de Nueva_columna de valores NaN, también se puede usar np.NaN
df_salaries
```

Mostrar el número de filas o columnas que tiene un DataFrame


```python
df_salaries.shape[0] #Muestra el numero de filas que posee el dataFrame
```

Agregar valores a una nueva columna


```python
data = np.arange(0, df_salaries.shape[0])
```

Crear una nueva columna y agregar los valores almacenados en el array


```python
df_salaries['Rango'] = data #Crea una nueva columna llamada Rango con los valores del array 
df_salaries
```

Para añadir filas se utiliza la función append de Python añadiendo como parámetro una lista, diccionario o añadiendo los valores manualmente.


```python
df_concat=pd.concat([df_salaries, df_salaries],sort=False,axis=0) 
```

axis =0 para agregarlo en filas y =1 para agregarlo en columnas, si ignore_index=True, agrega las casillas pero empezando desde el índice en el que se quedó.

También es posible usar *df_salaries.append(df_salaries)*, el cual duplica las filas del dataFrame, pero actualmente este método es deprecated

```python
df_concat.shape
```

## Manejo de datos nulos

Los datos nulos son dolores de cabeza para este mundo de la ciencia de datos y se van a encontrar mucho en nuestros DataFrames


Creamos un DataFrame con algunos valores nulos


```python

dict = {'Col1':[1,2,3,np.nan,5],
'Col2':[6, np.nan,8,9,pd.NA],
'Col3':['a','b','c', None, "e"]}
df = pd.DataFrame(dict)
df
```


Identificar valores nulos en un DataFrame


```python
df.isnull()
```


Identificar valores nulos con un valor numérico


```python
df.isnull()*1
```


Sustituir los valores nulos por una cadena


```python
df.fillna('Missing')
```

Sustituir valores nulos por una medida estadística realizada con los valores de las columnas


```python
df.fillna(df.mean())
```

Sustituir valores nulos por valores de interpolación


```python
df.interpolate()
```


Eliminar valores nulos


```python
df.dropna()
```

## Filtrado por condiciones

Funciona por lógica booleana y retorna los valores que están en “True”. Es muy útil porque en ocasiones queremos filtrar o separar datos.


```python
df_salaries.head()
```

Mostrar datos que sean mayores a cierto valor


```python
mayor40000 = df_salaries['salary'] >= 40000
mayor40000   #muestra el dataFrame con valores booleanos. True para salarios mayores a 40000
```


Filtrar datos en nuestro DataFrame que sean mayores a cierto valor


```python
df_salaries[mayor40000]  #filtra los datos que cumplen con la condicion
```

También se puede colocar la condición directamente como parámetro


```python
df_salaries[df_salaries['salary'] >= 40000] #filtra los datos que cumplen con la condicion
```

Mostrar los datos que sean igual a cierto valor


```python
genreFiction = df_salaries['experience_level'] == 'MI'
genreFiction    #muestra el dataFrame con valores booleanos. True para libros de tipo Fiction
```

Filtrado con varias condiciones


```python
df_salaries[genreFiction & mayor40000]   #Filtra los libros que sean de tipo Fiction y que hayan sido publicado desde 2017
```

Filtrado con negación


```python
df_salaries[~mayor40000] #Filtra los libros publicados antes o igual al 2016
```

También es posible hacer filstros de texto aplicando la función *str.contain()* a la lista o columna del dataframe que queremos filtrar

```python
scientist=df_salaries["job_title"].str.contains("Scientist")
scientist
```

```python
df_salaries[scientist]
```
